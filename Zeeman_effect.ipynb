$$ \text{Equation for gaussian fit: }f(x) = Ae^{-\frac{(x-B)^2}{2C^2}} + D$$

$$ \text{Equation for multiple gaussian fits: }f(x) = A_{1}e^{-\frac{(x-B_{1})^2}{2C_{1}^2}} + A_{2}e^{-\frac{(x-B_{2})^2}{2C_{2}^2}} + \text{ . . . } + D$$

Where $A$ is the amplitude/count of the peak, $B$ is the mean/channel number, $C$ is the standard deviation/width of the gaussian fit, and $D$ is a vertical offset.  

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.stats as stats

def fgaussian(x, A_1, B_1, C_1,
             A_2, B_2, C_2,
             A_3, B_3, C_3,
             A_4, B_4, C_4,
             A_5, B_5, C_5,
             A_6, B_6, C_6,
             A_7, B_7, C_7,
             A_8, B_8, C_8,
             A_9, B_9, C_9, D):
    return ((A_1 * np.exp(-((x - B_1) ** 2) / (2 * C_1 ** 2)))+(A_2 * np.exp(-((x - B_2) ** 2) / (2 * C_2 ** 2)))+
            (A_3 * np.exp(-((x - B_3) ** 2) / (2 * C_3 ** 2)))+(A_4 * np.exp(-((x - B_4) ** 2) / (2 * C_4 ** 2)))+
            (A_5 * np.exp(-((x - B_5) ** 2) / (2 * C_5 ** 2)))+(A_6 * np.exp(-((x - B_6) ** 2) / (2 * C_6 ** 2)))+
            (A_7 * np.exp(-((x - B_7) ** 2) / (2 * C_7 ** 2)))+(A_8 * np.exp(-((x - B_8) ** 2) / (2 * C_8 ** 2)))+
            (A_9 * np.exp(-((x - B_9) ** 2) / (2 * C_9 ** 2)))+D)
            

def is_float(string): 
    try:
        float(string)
        return True
    except ValueError:
        return False


def fit_and_plot_gaussian(x_data, y_data, initial_guess):

    # Perform the curve fitting
    params, covariance = curve_fit(fgaussian, x_data, y_data, p0=initial_guess)
    (A_1_fit, B_1_fit, C_1_fit,
    A_2_fit, B_2_fit, C_2_fit,
    A_3_fit, B_3_fit, C_3_fit,
    A_4_fit, B_4_fit, C_4_fit,
    A_5_fit, B_5_fit, C_5_fit,
    A_6_fit, B_6_fit, C_6_fit,
    A_7_fit, B_7_fit, C_7_fit,
    A_8_fit, B_8_fit, C_8_fit,
    A_9_fit, B_9_fit, C_9_fit, D_fit) = params

    uncertainties = np.sqrt(np.diag(covariance))

    # Plotting the data and the fit
    plt.figure(figsize=(8, 6))
    plt.xlabel('Wavelength')  
    plt.ylabel('Current')
    plt.plot(x_data, y_data, 'ro', ms=6, label='Data')
    plt.plot(x_data, fgaussian(x_data, *params), 'b--', linewidth=2, label='Best Fit')
    plt.legend()
    plt.show()

    # Print the fitting parameters
    print(f'Fit Parameters:\nA = {A_1_fit:.4f} ± {uncertainties[0]:.4f}\n'
         f'B_1 = {B_1_fit:.4f} ± {uncertainties[1]:.4f}\n'
         f'C_1 = {C_1_fit:.4f} ± {uncertainties[2]:.4f}\n\n'
         
         f'A_2 = {A_2_fit:.4f} ± {uncertainties[3]:.4f}\n'
         f'B_2 = {B_2_fit:.4f} ± {uncertainties[4]:.4f}\n'
         f'C_2 = {C_2_fit:.4f} ± {uncertainties[5]:.4f}\n\n'
         
         f'A_3 = {A_3_fit:.4f} ± {uncertainties[6]:.4f}\n'
         f'B_3 = {B_3_fit:.4f} ± {uncertainties[7]:.4f}\n'
         f'C_3 = {C_3_fit:.4f} ± {uncertainties[8]:.4f}\n\n'

         f'A_4 = {A_4_fit:.4f} ± {uncertainties[9]:.4f}\n'
         f'B_4 = {B_4_fit:.4f} ± {uncertainties[10]:.4f}\n'
         f'C_4 = {C_4_fit:.4f} ± {uncertainties[11]:.4f}\n\n'
         
         f'A_5 = {A_5_fit:.4f} ± {uncertainties[12]:.4f}\n'
         f'B_5 = {B_5_fit:.4f} ± {uncertainties[13]:.4f}\n'
         f'C_5 = {C_5_fit:.4f} ± {uncertainties[14]:.4f}\n\n'
         
         f'A_6 = {A_6_fit:.4f} ± {uncertainties[15]:.4f}\n'
         f'B_6 = {B_6_fit:.4f} ± {uncertainties[16]:.4f}\n'
         f'C_6 = {C_6_fit:.4f} ± {uncertainties[17]:.4f}\n\n'
         
         f'A_7 = {A_7_fit:.4f} ± {uncertainties[18]:.4f}\n'
         f'B_7 = {B_7_fit:.4f} ± {uncertainties[19]:.4f}\n'
         f'C_7 = {C_7_fit:.4f} ± {uncertainties[20]:.4f}\n\n'
         
         f'A_8 = {A_8_fit:.4f} ± {uncertainties[21]:.4f}\n'
         f'B_8 = {B_8_fit:.4f} ± {uncertainties[22]:.4f}\n'
         f'C_8 = {C_8_fit:.4f} ± {uncertainties[23]:.4f}\n\n'
         
         f'A_9 = {A_9_fit:.4f} ± {uncertainties[24]:.4f}\n'
         f'B_9 = {B_9_fit:.4f} ± {uncertainties[25]:.4f}\n'
         f'C_9 = {C_9_fit:.4f} ± {uncertainties[26]:.4f}\n\n'

         f'D = {D_fit:.4f} ± {uncertainties[27]:.4f}')

datam = np.genfromtxt('Data/data_zeeman_high_field1.csv', delimiter=',', dtype=str, encoding='utf-8-sig')

# split columns from data into x and y values 
x_data_m = [float(row[0]) if is_float(row[0]) else np.nan for row in datam]
y_data_m = [float(row[1]) if is_float(row[1]) else np.nan for row in datam]

```

#### NEXT STEPS:

- Take the wavelength of each peak and solve for energy $E_i = \frac{hc}{\lambda_i}$
- Calculate the change in energy of the photon $\Delta E = E_i - E_0$, where $E_0$ is the energy of the `center line`
- Create a linear fit using $\Delta E$ and values of n, n ranges from -4 to 4.
- we are looking for an equation that looks like this, $\Delta E = c\mu_bB n$, where our slope,m, in the fit should be $m=c\mu_bB$
- c is 1/2, $\mu_b$ is bohr magneton, and B is magnetic filed of 1.3ish Tesla.
- We then want to use our slope to calculate the bohr magneton and compare to its value on google.














```python
E_data = np.array([0.081, 0.088, 0.122, 0.276, 
                   0.303, 0.356, 0.384, 0.511, 
                   0.835, 1.115, 1.175, 1.333])

C_data = np.array([155.16810774, 166.37276477, 226.98617824, 
                   479.72620940, 526.84697657, 613.82722775,
                   690.00512348, 845.18251807, 1281.95949070, 
                   1814.11727851, 1712.56931386, 1895.20371921]) 

C_errors = np.array([0.16372205, 0.21175686, 0.07422393, 
                  7.95216302, 0.47711423, 0.17468657, 
                  3.58939626, 0.20642624, 0.32791821, 
                  1.01625214, 1.77457388, 1.07130561])

slope, intercept, r_value, p_value, std_err = stats.linregress(E_data, C_data)

result = stats.linregress(E_data, C_data)

data_labels = ['Ba-133', 'Cd-109', 'Co-57', 
               'Ba-133', 'Ba-133', 'Ba-133',
               'Ba-133', 'Na-22', 'Mn-54', 
               'Na-22', 'Co-60', 'Co-60']

for x, y, label in zip(E_data, C_data, data_labels):
    plt.text(x, y, label, fontsize=9, ha='left', va='top')

res = stats.linregress(E_data, C_data)
#plt.plot(E_data, C_data, '.', color='blue', label='data points')
plt.plot(E_data, res.intercept + res.slope*E_data, 'r', label='fitted line')
plt.errorbar(E_data, C_data, yerr=C_errors, fmt='.', color='blue', 
             label='Data points w/ error in channel', markersize=5, capsize=3)
plt.xlabel('Energy [MeV]')
plt.ylabel('Channel')
plt.title('Energy vs. Counts for Gamma Ray Sources')
plt.grid(True)
plt.legend()
plt.show()

print(f"Linear fit: C = ({res.slope:.8f}+/-{res.stderr:.8f})E + ({res.intercept:.8f}+/-{res.stderr:.8f})")
```
