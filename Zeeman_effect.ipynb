$$ \text{Equation for gaussian fit: }f(x) = Ae^{-\frac{(x-B)^2}{2C^2}} + D$$

$$ \text{Equation for multiple gaussian fits: }f(x) = A_{1}e^{-\frac{(x-B_{1})^2}{2C_{1}^2}} + A_{2}e^{-\frac{(x-B_{2})^2}{2C_{2}^2}} + \text{ . . . } + D$$

Where $A$ is the amplitude/count of the peak, $B$ is the mean/channel number, $C$ is the standard deviation/width of the gaussian fit, and $D$ is a vertical offset.  

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.stats as stats

def fgaussian(x, A_1, B_1, C_1,
             A_2, B_2, C_2,
             A_3, B_3, C_3,
             A_4, B_4, C_4,
             A_5, B_5, C_5,
             A_6, B_6, C_6,
             A_7, B_7, C_7,
             A_8, B_8, C_8,
             A_9, B_9, C_9, D):
    
    return ((A_1 * np.exp(-((x - B_1) ** 2) / (2 * C_1 ** 2)))+
            (A_2 * np.exp(-((x - B_2) ** 2) / (2 * C_2 ** 2)))+
            (A_3 * np.exp(-((x - B_3) ** 2) / (2 * C_3 ** 2)))+
            (A_4 * np.exp(-((x - B_4) ** 2) / (2 * C_4 ** 2)))+
            (A_5 * np.exp(-((x - B_5) ** 2) / (2 * C_5 ** 2)))+
            (A_6 * np.exp(-((x - B_6) ** 2) / (2 * C_6 ** 2)))+
            (A_7 * np.exp(-((x - B_7) ** 2) / (2 * C_7 ** 2)))+
            (A_8 * np.exp(-((x - B_8) ** 2) / (2 * C_8 ** 2)))+
            (A_9 * np.exp(-((x - B_9) ** 2) / (2 * C_9 ** 2)))+D)
            

def is_float(string): 
    try:
        float(string)
        return True
    except ValueError:
        return False


def fit_and_plot_gaussian(x_data, y_data, initial_guess):

    # Perform the curve fitting
    params, covariance = curve_fit(fgaussian, x_data, y_data, p0=initial_guess, maxfev=5000)
    (A_1_fit, B_1_fit, C_1_fit,
    A_2_fit, B_2_fit, C_2_fit,
    A_3_fit, B_3_fit, C_3_fit,
    A_4_fit, B_4_fit, C_4_fit,
    A_5_fit, B_5_fit, C_5_fit,
    A_6_fit, B_6_fit, C_6_fit,
    A_7_fit, B_7_fit, C_7_fit,
    A_8_fit, B_8_fit, C_8_fit,
    A_9_fit, B_9_fit, C_9_fit, 
    D_fit) = params

    peaks = np.array([B_1_fit, B_2_fit, B_3_fit, B_4_fit, B_5_fit, B_6_fit, B_7_fit, B_8_fit, B_9_fit])
    uncertainties = np.sqrt(np.diag(covariance))
    b_uncertainties = uncertainties[1:-1:3]
    # Plotting the data and the fit
    plt.figure(figsize=(8, 6))
    plt.xlabel('Wavelength')  
    plt.ylabel('Current')
    plt.plot(x_data, y_data, 'ro', ms=6, label='Data')
    plt.plot(x_data, fgaussian(x_data, *params), 'b--', linewidth=2, label='Best Fit')
    plt.legend()
    plt.show()

    # Print the fitting parameters
    print(f'Fit Parameters:\nA_1 = {A_1_fit:.4f} ± {uncertainties[0]:.4f}\n'
         f'B_1 = {B_1_fit:.4f} ± {uncertainties[1]:.4f}\n'
         f'C_1 = {C_1_fit:.4f} ± {uncertainties[2]:.4f}\n\n'
         
         f'A_2 = {A_2_fit:.4f} ± {uncertainties[3]:.4f}\n'
         f'B_2 = {B_2_fit:.4f} ± {uncertainties[4]:.4f}\n'
         f'C_2 = {C_2_fit:.4f} ± {uncertainties[5]:.4f}\n\n'
         
         f'A_3 = {A_3_fit:.4f} ± {uncertainties[6]:.4f}\n'
         f'B_3 = {B_3_fit:.4f} ± {uncertainties[7]:.4f}\n'
         f'C_3 = {C_3_fit:.4f} ± {uncertainties[8]:.4f}\n\n'

         f'A_4 = {A_4_fit:.4f} ± {uncertainties[9]:.4f}\n'
         f'B_4 = {B_4_fit:.4f} ± {uncertainties[10]:.4f}\n'
         f'C_4 = {C_4_fit:.4f} ± {uncertainties[11]:.4f}\n\n'
         
         f'A_5 = {A_5_fit:.4f} ± {uncertainties[12]:.4f}\n'
         f'B_5 = {B_5_fit:.4f} ± {uncertainties[13]:.4f}\n'
         f'C_5 = {C_5_fit:.4f} ± {uncertainties[14]:.4f}\n\n'
         
         f'A_6 = {A_6_fit:.4f} ± {uncertainties[15]:.4f}\n'
         f'B_6 = {B_6_fit:.4f} ± {uncertainties[16]:.4f}\n'
         f'C_6 = {C_6_fit:.4f} ± {uncertainties[17]:.4f}\n\n'
         
         f'A_7 = {A_7_fit:.4f} ± {uncertainties[18]:.4f}\n'
         f'B_7 = {B_7_fit:.4f} ± {uncertainties[19]:.4f}\n'
         f'C_7 = {C_7_fit:.4f} ± {uncertainties[20]:.4f}\n\n'
         
         f'A_8 = {A_8_fit:.4f} ± {uncertainties[21]:.4f}\n'
         f'B_8 = {B_8_fit:.4f} ± {uncertainties[22]:.4f}\n'
         f'C_8 = {C_8_fit:.4f} ± {uncertainties[23]:.4f}\n\n'
         
         f'A_9 = {A_9_fit:.4f} ± {uncertainties[24]:.4f}\n'
         f'B_9 = {B_9_fit:.4f} ± {uncertainties[25]:.4f}\n'
         f'C_9 = {C_9_fit:.4f} ± {uncertainties[26]:.4f}\n\n'

         f'D = {D_fit:.4f} ± {uncertainties[27]:.4f}')
    
    return peaks, b_uncertainties

data = np.genfromtxt('Data.csv', delimiter=',', dtype=str, encoding='utf-8-sig', skip_header=1)

# split columns from data into x and y values 
x_data_best = [float(row[0]) if is_float(row[0]) else np.nan for row in data]
y_data_best = [float(row[1]) if is_float(row[1]) else np.nan for row in data]
x_data_polarL = [float(row[2]) if is_float(row[2]) else np.nan for row in data]
y_data_polarL = [float(row[3]) if is_float(row[3]) else np.nan for row in data]
x_data_polarR = [float(row[4]) if is_float(row[4]) else np.nan for row in data]
y_data_polarR = [float(row[5]) if is_float(row[5]) else np.nan for row in data]
```

### Fit of 9 peaks

```python
initial_guess = [0.1, 545.870, 0.006,   # Peak 1
                 0.15, 545.880, 0.005,  # Peak 2
                 0.2, 545.883, 0.005,   # Peak 3
                 0.28, 545.899, 0.01,   # Peak 4
                 0.27, 545.911, 0.008,  # Peak 5
                 0.19, 545.927, 0.001,  # Peak 6
                 0.16, 545.935, 0.001,  # Peak 7
                 0.13, 545.943, 0.001,  # Peak 8
                 0.10, 545.955, 0.006,  # Peak 9
                 0.40]                  # Baseline

peaks, uncertainties = fit_and_plot_gaussian(np.array(x_data_best), np.array(y_data_best), initial_guess)
```

```python
h = 6.62607015e-34  # Planck's constant in J*s
c = 299792458       # Speed of light in m/s
B = 1.3             # Magnetic field strength in Tesla
C = 1/2        # Constant C (1/2)
mu_b_true = 9.274009994e-24  # True value of Bohr magneton
n_values = np.arange(-4, 5)  # Values of n from -4 to 4

# Calculate energy for each peak
energies = h * c / (peaks * 1e-9)
energy_uncertainties = h * c / (uncertainties * 1e-9)
print (energy_uncertainties)
# Calculate delta E
centerline_energy = energies[4]
delta_E = [E_i - centerline_energy for E_i in energies]

# Perform linear fit
def linear_fit(x, m):
    return m * x

params, covariance = curve_fit(linear_fit, n_values, delta_E)
slope, slope_error = params[0], np.sqrt(covariance[0, 0])
print(slope)
print (slope_error)

# Calculate Bohr magneton
mu_b = slope / (C * B)
mu_b_error = mu_b * np.sqrt((slope_error / slope)**2)

# Print results
print("Slope (m): {:.4e} +/- {:.4e}".format(-slope, slope_error))
print("Bohr magneton (mu_b): {:.4e} +/- {:.4e}".format(-mu_b, mu_b_error))

# Compare to true value
error = abs(mu_b - mu_b_true) / mu_b_true
print("Relative error compared to true value: {:.4f}".format(error))

# Plot data and linear fit
plt.figure(figsize=(8, 6))
plt.errorbar(-1 * n_values, delta_E, yerr=(1e-11)*energy_uncertainties, fmt='ro', ms=6, label='Data with Error Bars')
plt.plot(n_values, linear_fit(-1*n_values, slope), color='blue', label='Linear Fit')
plt.xlabel('n')
plt.ylabel('Delta E')
plt.title('Linear Fit of Delta E vs. n')
plt.legend()
plt.grid(True)
plt.show()
```
